```python
#Reading and storing the data to be analysed in the variable 'data'
data = pd.read_csv('data_clean/clean_data.csv')
```

```python
#Task-c:

encoded_data = pd.get_dummies(data, columns=['Fuel_Type', 'Transmission'])
encoded_data.head(10)
```

```python
# Dropping the first column which is just a row number
encoded_data = encoded_data.iloc[:, 1:]

# Write encoded data to a new file
encoded_data.to_csv('/content/drive/MyDrive/PDS_Assignment_2/results/encoded_data.csv', index=False)

encoded_data.head(10)
```

```python
#Task-d:

import datetime as dt
current_year = dt.datetime.now().year
data['Age'] = current_year - data['Year']

data = data.iloc[:, 1:]

# Write the data of cars age to a new file
data.to_csv('/content/drive/MyDrive/PDS_Assignment_2/results/cars_age.csv', index=False)

data.head(10)
```

```python
#Task-e:

# Select: Select specific columns
selected_data = data[['Name', 'Location', 'Year', 'Mileage', 'Fuel_Type', 'Transmission']]

# Filter: Filter rows based on conditions
filtered_data = data[data['Price'] > data['Price'].mean()]

# Rename: Rename columns
renamed_data = data.rename(columns={'Year': 'Manufacture_Year', 'Price': 'Selling_Price'})

# Mutate
data['Price_Category'] = pd.cut(data['Price'], bins=[0, 5, 10, 15, 20, 25, 30], labels=['<5', '5-10', '10-15', '15-20', '20-25', '>25'])

# Arrange: Sort rows based on one or more columns
arranged_data = data.sort_values(by='Price', ascending=False)

# Summarize with group by :
# Finding the average price of cars based on their location
summary_data = data.groupby('Location')['Price'].mean()

# Results
print("Select:")
print(selected_data.head())

print("----------------------")
print("\nFilter:")
print(filtered_data.head())


print("----------------------")
print("\nRename:")
print(renamed_data.head())


print("----------------------")
print("\nMutate:")
print(data.head())


print("----------------------")
print("\nArrange:")
print(arranged_data.head())


print("----------------------")
print("\nSummary Data (Average Price by Location):")
print(summary_data)
```
