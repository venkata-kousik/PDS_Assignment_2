```python
data = pd.read_csv('data_raw/raw_data.csv')
data
```

```python
# Task-b: Convert Mileage, Engine, Power and New_Price to numerical values by removing units
data['Mileage'] = data['Mileage'].str.split().str[0].astype(float)
data['Engine'] = data['Engine'].str.split().str[0].astype(float)
data['Power'] = data['Power'].str.split().str[0].astype(float)
data['New_Price'] = data['New_Price'].str.split().str[0].astype(float)
```

```python
data.head(5)
```

```python
#Task-a:

# Checking for the missing values
missing_values = data.isnull().sum()
missing_values
```

```python
# Drop unnecessary columns
data.drop(columns=['New_Price'], inplace=True)

# Handle missing values
data['Mileage'].fillna(data['Mileage'].mean(), inplace=True)
data['Engine'].fillna(data['Engine'].mean(), inplace=True)
data['Power'].fillna(data['Power'].mean(), inplace=True)
data['Seats'].fillna(data['Seats'].mean(), inplace=True)
```

```python
# Checking if the dataset still has any null values
print(data.isnull().sum())
```

```python
data.head(10)
```

```python
# Dropping the first column which is just a row number
data = data.iloc[:, 1:]
```
